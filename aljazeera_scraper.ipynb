```python
df = pd.read_csv('final_cleaned_dataset.csv')
```

```python
df2 = df.loc[df['Source']=='Al Jazeera English']
df2.head()
```

```python
d_l = []
count = 0
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
for url in df['URL']:
    browser.visit(url)
    time.sleep(1)
    html = browser.html   
    soup = BeautifulSoup(html, 'html.parser')
#     article = soup.find('div', class_='wysiwyg wysiwyg--all-content')
    body = ''
    x = soup.find_all('p', class_=False)
    for one in x:
        body += one.text + '\n'
    dictionary = {
        'url': url,
        'body': body,
    }
    d_l.append(dictionary)
    time.sleep(1)
    count +=  1
    print(count)

browser.quit()
```
