```python
from bs4 import BeautifulSoup
import html
from splinter import Browser
import pandas as pd
import time
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.foxnews.com/politics/michigan-lawmakers-meet-with-trump'
browser.visit(url)
```

```python
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
tags = soup.find_all('p', class_=False)
tags
```

```python
for tag in tags:
    tag['class'] = 'speakable'
```

```python
soup.find_all('p', class_='speakable')
```

```python
df = pd.read_csv('final_cleaned_dataset.csv')
```

```python
df5 = df.loc[df['Source']=='Fox News']
df5.head()
```

```python
count = 0
d_l = []
```

```python
for url in df5['URL']:
    browser.visit(url)
    time.sleep(1)
    html = browser.html   
    soup = BeautifulSoup(html, 'html.parser')
    body = ''
    tags = soup.find_all('p', class_=False)
    for tag in tags:
        tag['class'] = 'speakable'
    x = soup.find_all('p', class_='speakable')
    for one in x:
        body += one.text + '\n'
    dictionary = {
        'url': url,
        'body': body,
    }
    d_l.append(dictionary)
    count +=  1
    print(count)

browser.quit()
```
