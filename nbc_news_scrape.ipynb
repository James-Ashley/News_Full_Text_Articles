```python
from bs4 import BeautifulSoup
import html
from splinter import Browser
import pandas as pd
import time
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.nbcnews.com/politics/2020-election/federal-judge-dismisses-trump-election-lawsuit-pennsylvania-n1248528'
browser.visit(url)
```

```python
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
text = soup.find_all('p', class_='endmarkEnabled')
for x in text:
    print(x.text)
```

```python
df = pd.read_csv('final_cleaned_dataset.csv')
```

```python
df4 = df.loc[df['Source']=='NBC News']
df4.head()
```

```python
count = 0
d_l = []
```

```python
for url in df4['URL']:
    browser.visit(url)
    time.sleep(1)
    html = browser.html   
    soup = BeautifulSoup(html, 'html.parser')
    body = ''
    x = soup.find_all('p', class_='endmarkEnabled')
    for one in x:
        body += one.text + '\n'
    dictionary = {
        'url': url,
        'body': body,
    }
    d_l.append(dictionary)
    count +=  1
    print(count)

browser.quit()
```

```python
for story in d_l:
    if story['body'] == '':
        print(story['url'])
```
